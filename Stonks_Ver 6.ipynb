## Table: Stock Price Data Obtained From AlphaVantage

For TA's information,

**Open** is the price of the stock when the market has **started that day**. (Usually 9.30AM ET)

**High** is the **highest price** the stock managed to reach that day.

**Low** is the **lowest price** the stock managed to reach that day.

```python
# We load up all the required files here.
# We clean up the data by taking only the data we need, creating a dataframe for every indicator collected, 
# and only taking the first 2000 points. 

with open("Data\Stock Time Series Data\AAPL4.json") as json_file: 
    jsonStock = json.load(json_file)
stockPrice = pd.DataFrame(jsonStock["Time Series (Daily)"],["5. adjusted close"]).T.iloc[6:2006]
stockPrice = stockPrice.rename(columns={"5. adjusted close" : 'Close'})

with open("Data\Indicators\SMA10.json") as json_file: 
    jsonSMA10 = json.load(json_file)
SMA10 = pd.DataFrame(jsonSMA10["Technical Analysis: SMA"]).T.iloc[:2000]

with open("Data\Indicators\EMA12.json") as json_file: 
    jsonEMA12 = json.load(json_file)
EMA12 = pd.DataFrame(jsonEMA12["Technical Analysis: EMA"]).T.iloc[:2000]

with open("Data\Indicators\WMA10.json") as json_file: 
    jsonWMA10 = json.load(json_file)
WMA10 = pd.DataFrame(jsonWMA10["Technical Analysis: WMA"]).T.iloc[:2000]

with open("Data\Indicators\DEMA10.json") as json_file: 
    jsonDEMA10 = json.load(json_file)
DEMA10 = pd.DataFrame(jsonDEMA10["Technical Analysis: DEMA"]).T.iloc[:2000]

with open("Data\Indicators\TEMA10.json") as json_file: 
    jsonTEMA10 = json.load(json_file)
TEMA10 = pd.DataFrame(jsonTEMA10["Technical Analysis: TEMA"]).T.iloc[:2000]

with open("Data\Indicators\TRIMA10.json") as json_file: 
    jsonTRIMA10 = json.load(json_file)
TRIMA10 = pd.DataFrame(jsonTRIMA10["Technical Analysis: TRIMA"]).T.iloc[:2000]

with open("Data\Indicators\KAMA10.json") as json_file: 
    jsonKAMA10 = json.load(json_file)
KAMA10 = pd.DataFrame(jsonKAMA10["Technical Analysis: KAMA"]).T.iloc[:2000]

with open("Data\Indicators\MAMA10.json") as json_file: 
    jsonMAMA10 = json.load(json_file)
MAMA10 = pd.DataFrame(jsonMAMA10["Technical Analysis: MAMA"]).T.iloc[:2000]
```

```python
# We combine the collected data here for us to easily see the data of each technical indicator.

combine = [stockPrice, SMA10, EMA12, WMA10, DEMA10, TEMA10, TRIMA10, KAMA10, MAMA10]

result = pd.concat(combine, axis=1).astype(float)
result = result.iloc[::-1]

result['PercentageChange'] = pd.Series(dtype ='float')
result['PercentageChange'] = result['Close'].pct_change()

result['Class'] = pd.Series(dtype ='string')

for i in range(2000):
    if result['PercentageChange'][i] > 0 : result['Class'][i] = 'Up'
    else: result['Class'][i] = 'Down'
        
result['Class'] = result['Class'].astype("category")

result.head()
```

We combined all of our Indicators and ClosingPrice data into a dataframe, in order to analyse them easily.

> PercentageChange calculates the **change in closing price compared to the day before**. Therefore, our first row is NaN, as there is no previous reference that it can refer to.

Class is a **categorical** column derived from PercentageChange:
> If the PercentageChange is **positive**: Class will categorise it as **Up**, which means the **closing price has increased** in comparison to the day before.  
> If the PercentageChange is **negative**: Class will categorise it as **Down**, which means the **closing price has decreased** in comparison to the day before.

# 1st Approach: Linear Regression Model

```python
# d = predictor, Moving Averages.
# e = response, close price.

e = pd.DataFrame(result['Close']) # Response
for counter, item in enumerate(MA):
    d = pd.DataFrame(MA[item]) # We need to pass a dataframe instead of MA[item] directly
    
    d_train, d_test, e_train, e_test = train_test_split(d, e, test_size = 360)
    
    linreg.fit(d_train,e_train) 
    
    e_train_pred = linreg.predict(e_train)
    
    print('====================================================')
    print()
    print('Relation between Close and', item)
    print()
    print('Intercept \t: b = ', linreg.intercept_)
    print('Coefficients: a = ', linreg.coef_)
    print()
    print("Explained Variance (R^2) \t:", linreg.score(d_train, e_train))
    print("Mean Squared Error (MSE) \t:", mean_squared_error(e_train, e_train_pred))
    print()
    jointDF = pd.concat([e, d], axis = 1).reindex(e.index)
    sb.jointplot(data = jointDF, x = item, y = "Close", height = 10, kind = "reg", joint_kws={'line_kws':{'color':'red'}}, space = 0)
```

Based on the Linear Regression Model above, we can see that the model took into account that the Close Price are all averaged out.

The beauty of linear regression is that the stock's price and time period determine the system parameters making it universally applicable. However, having only 1 regression line does not really tells us the following:
> i. Entry Point:  The price at which an investor buys or sells a security.  
> ii. Stop-loss Point:  Defined as an advance order to sell an asset when it reaches a particular price point. It is used to limit loss or gain in a trade.  
> iii. Profitable exits: An exit point is the price at which an investor or trader closes a position

Futhermore, with how **close or similar** the value of the R^2 and MSE that the majority of our MA indicators have, it further hinders our attempt in determining the best MA indicator.

As such, the Linear Regression Model does not sufficiently help us sufficiently.

Therefore, this results in us having to look for an alternative that is able to further separate each of the MA indicators.

# 2nd Approach: Decision Tree Model

Moving on, we proceed to split the data into 2 parts:

> Train Data 70%
>
> Test Data 30%

### Visualising how our data will look like after training them using boxplots

```python
# Storage for classification accuracy
dectreescore_train = []
dectreescore_test = []

# Storage for confusion matrix
cm_train = []
cm_test = []

# Storage for dectree
dectreestore = []
featurename = []

CT_response = pd.DataFrame(result['Class']) # Categorical  

for counter, item in enumerate(MA):
    CT_predictor = pd.DataFrame(MA[item]) # We need to pass a dataframe for iteration to work
    CT_predictor_train, CT_predictor_test, CT_response_train, CT_response_test = train_test_split(
        CT_predictor, CT_response, test_size = 0.25)
    
    dfTrain = pd.concat([CT_response_train,CT_predictor_train], axis = 1).reindex(CT_response_train.index)
    
    # Plotting all the boxplot to understand how the concentration will look like
    f, axes = plt.subplots(1,1, figsize=(18, 9))
    sb.boxplot(x = item, y = "Class", data = dfTrain, orient = "h",)   
    
    # Decision Tree using Train Data, we will need to: 
    # i. Create a decision tree model
    # ii. Training the model created
    dectree = DecisionTreeClassifier(max_depth = 3)
    dectree.fit(CT_predictor_train, CT_response_train)

    # Predict Response corresponding to Predictors
    CT_response_train_pred = dectree.predict(CT_predictor_train) 
    CT_response_test_pred = dectree.predict(CT_predictor_test)

    # Store the appended to perform a loop to print the Decision Tree Model later.
    dectreestore.append(dectree.fit(CT_predictor_train, CT_response_train))
    featurename.append(CT_predictor_train.columns)
    
    # Store the values to perform a loop to print the Classification Accuracy later
    dectreescore_train.append(dectree.score(CT_predictor_train, CT_response_train))
    dectreescore_test.append(dectree.score(CT_predictor_test, CT_response_test))
    
    # Store the values to perform a loop to print the Confusion Matrix later
    cm_train.append(confusion_matrix(CT_response_train, CT_response_train_pred))
    cm_test.append(confusion_matrix(CT_response_test, CT_response_test_pred))
```

### Visualising how our Decision Tree Model will look like after training

```python
for counter, item in enumerate(MA):

    f = plt.figure(figsize=(16,8))
    plot_tree(dectreestore[counter], filled=True, rounded=True, 
              feature_names=featurename[counter], 
              class_names=["Up","Down"])
```

### Print all the rates from the Confusion Matrix

The 2 most important factor are **TPR and TNR** that will affect our decision in determining the best MA indicator.

In the context of market trading:
> Accuracy  How often is the classifier correct? TP+TN/Total   
> Precision  How often are the buy predictions correct? TP/TP+FP  
> Sensitivity  Also called True Positive Rate. What percentage of the buy labels was the classifier able to identify?  
> Specificity  Also called True Negative Rate. What percentage of the sell (0) labels were identified correctly by the classifier?   
> Misclassification Rate  How often is the classifier wrong? FP + FN / Total  
> Null Error Rate - How often you would be wrong if you always predicted the majority class.

In order to generalise Null Error, we will assume that we are always predicting the True Close Price: FN + TP / Total

| Confusion Matrix  |       |        |        |      
| :---              | :---: | :----: | :----: |         
| Actual Negative   |  (0)  |   TN   |   FP   |             
| Actual Positive   |  (1)  |   FN   |   TP   |       
|                   |       |   (0)   |   (1)   |       
|                   |       | Predicted Negative    |   Predicted Postitive  |     


* `TPR = TP / (TP + FN)` : True Positive Rate = True Positives / All Positives    
* `TNR = TN / (TN + FP)` : True Negative Rate = True Negatives / All Negatives    

* `FPR = FP / (TN + FP)` : False Positive Rate = False Positives / All Negatives 
* `FNR = FN / (TP + FN)` : False Negative Rate = False Negatives / All Positives 

```python
for counter, item in enumerate(MA):
    
    cmTrain = cm_train[counter]
    tpTrain = cmTrain[1][1] # True Positives : Y (1) predicted Y (1)
    fpTrain = cmTrain[0][1] # False Positives : N (0) predicted Y (1)
    tnTrain = cmTrain[0][0] # True Negatives : N (0) predicted N (0)
    fnTrain = cmTrain[1][0] # False Negatives : Y (1) predicted N (0)    

    cmTest = cm_test[counter]
    tpTest = cmTest[1][1] # True Positives : Y (1) predicted Y (1)
    fpTest = cmTest[0][1] # False Positives : N (0) predicted Y (1)
    tnTest = cmTest[0][0] # True Negatives : N (0) predicted N (0)
    fnTest = cmTest[1][0] # False Negatives : Y (1) predicted N (0)
    
    acc_train = (tpTrain+tnTrain) / 1500
    pre_train = tpTrain / (tpTrain+fpTrain)
    sensi_train = tpTrain/(tpTrain + fnTrain)
    speci_train = tnTrain/(tnTrain + fpTrain)
    mis_train = (fpTrain+fnTrain)/1500
    null_train = (fnTrain+tpTrain) / 1500
    
    acc_test = (tpTest+tnTest) / 500
    pre_test = tpTest / (tpTest+fpTest)
    sensi_test = tpTest/(tpTest + fnTest)
    speci_test = tnTest/(tnTest + fpTest)
    mis_test = (fpTest+fnTest)/ 500
    null_test = (fnTest+tpTest) / 500
    
    print('================================')
    print('', item)
    print()
    print(" Train Dataset")
    print(" The Accuracy rate is: \t ", acc_train)
    print(" The Precision rate is: \t ", pre_train)
    print(" The Sensitivity rate is: \t ", sensi_train)
    print(" The Specificity rate is: \t ", speci_train)
    print(" The Misclassification rate is:", mis_train)
    print(" The Null Error rate is: \t ", null_train)
    print()    
    print(" Test Dataset")
    print(" The Accuracy rate is: \t ", acc_test)
    print(" The Precision rate is: \t ", pre_test)
    print(" The Sensitivity rate is: \t ", sensi_test)
    print(" The Specificity rate is: \t ", speci_test)
    print(" The Misclassification rate is:", mis_test)
    print(" The Null Error rate is: \t ", null_test)
    print()
```
